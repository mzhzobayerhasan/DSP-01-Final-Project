```python
train_set['Hour Of Day'].astype(float)
num_cols = train_set.select_dtypes(include=[np.number])
cat_cols = train_set.select_dtypes('object')

cat_cols
```

```python
num_cols
```

```python
from sklearn.preprocessing import StandardScaler

num_cols_numeric = num_cols.select_dtypes(include=['int64', 'float64'])

std_scaler = StandardScaler()
num_train_data = std_scaler.fit_transform(num_cols_numeric)

num_train_data_df = pd.DataFrame(num_train_data, columns=num_cols_numeric.columns)

num_train_data_df.head()
```

```python
num_train_data_df.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

onehot_encoder = OneHotEncoder()  # Updated sparse_output argument
ordinal_encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High']])  # For Traffic Density
```

```python
cat_cols['Traffic Density'] = ordinal_encoder.fit_transform(cat_cols[['Traffic Density']])
```

```python
import pandas as pd

# Assuming cat_cols is a DataFrame
data_encoded = pd.get_dummies(cat_cols.drop('City', axis=1), drop_first=True)

# Concatenate the encoded columns with 'Traffic Density'
cat_cols_encoded = pd.concat([data_encoded, cat_cols['Traffic Density']], axis=1)
cat_cols_encoded = cat_cols_encoded.astype(int)
```

```python
cat_cols_encoded.isnull().sum()
```

```python
num_train_data_df.shape
```

```python
cat_cols_encoded.shape
```

```python
num_cols_train_data_df = num_train_data_df.reset_index(drop=True)
cat_cols_encoded = cat_cols_encoded.reset_index(drop=True)
train_final_data = pd.concat([num_cols_train_data_df, cat_cols_encoded], axis=1)
train_final_data.shape
```

```python
train_final_data.isnull().sum()
```

```python
X = train_final_data.drop(['Energy Consumption', 'Traffic Density'], axis=1)
```

```python
y1=train_final_data['Energy Consumption']
y2=train_final_data['Traffic Density']
```

```python
X_t,X_v,y1_t,y1_v,y2_t,y2_v=train_test_split(X,y1,y2,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_t, y1_t)
```

```python
from sklearn.ensemble import RandomForestRegressor
RF = RandomForestRegressor(n_estimators=100, random_state=42)
RF.fit(X_t, y1_t)
```
